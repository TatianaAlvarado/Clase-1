# **4. Indexing**


```python
# Create an index
# ESTO NO ES FUNCIONAL DADO QUE LOS INDICES YA ESTAN EN LA BASE
idx = pd.MultiIndex.from_product([dd['Territory'].unique(),
                                 range(dd.Year.min(),dd.Year.max()+1)],
                                 names = ['Territory' , 'Year'])
# Reindex the dataframe
dd = dd.set_index(['Territory','Year'])

# Check
dd.head
```
